```python
# Carrega o dataset 
ds = load_dataset(
    "c01dsnap/CIC-IDS2017",
    split="train",
    streaming=True
)

rows = []
for i, row in enumerate(ds):
    if i >= 300000:
        break
    # Filtra apenas registros do dia Wednesday (ajuste o nome da coluna conforme necess√°rio)
    if row.get("Day") == "Wednesday":
        rows.append(row)

# Converte para DataFrame
df = pd.DataFrame(rows)

# Inspeciona as primeiras linhas
print(df.head())
```
